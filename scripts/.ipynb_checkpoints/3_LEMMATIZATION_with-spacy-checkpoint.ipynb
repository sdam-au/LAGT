# Preprocessing

```python
# short messy string for testing
test_string = "ἀρχόμενος ³σέο, <Φοῖβε>, text in latin also CAPITAL     ⁴παλαιγενέων 1234 κλέα 34056 φωτῶν μνήσομαι, οἳ Πόντοιο κατὰ στόμα καὶ διὰ πέτρας Κυανέας βασιλῆος ἐφημοσύνῃ Πελίαο χρύσειον μετὰ κῶας ἐύζυγον ἤλασαν Ἀργώ. τοίην γὰρ Πελίης φάτιν ἔκλυεν, ὥς μιν ὀπίσσω μοῖρα μένει στυγερή, τοῦδʼ ἀνέρος, ὅντινʼ ἴδοιτο δημόθεν οἰοπέδιλον, ὑπʼ ἐννεσίῃσι δαμῆναι."
test_string
```

```python
# remove numerical superscripts
def remove_sups(string):
    sups = ["¹", "²", "³", "⁴","⁵", "⁶", "⁷", "⁰", "⁹"]
    for sup in sups:
        string = string.replace(sup, "")
    return string
def normalize_encoding(string):
    return unicodedata.normalize("NFC", string)

def clean_string(string):
    string = normalize_encoding(string)
    string = remove_sups(string)
    string = re.sub("[a-zA-Z0-9]", "", string)
    string = re.sub("<|>", "", string)
    string = re.sub("\s+", " ", string)
    return string
```

```python
# checking how it works
clean_string(test_string)
```

```python
%%time
# new column for clean text
AGT["clean_string"] = AGT["string"].apply(clean_string)
```

```python
# add column with sentences
def get_sentences(string):
    splits = re.split(r"(\.|\·|\:|\;|\?|\!)" , string)
    sentences = []
    for n in range(0, len(splits), 2):
        try:
            sentence = splits[n]+splits[n+1]
            sentence = re.sub("^\s*", "", sentence)
            sentences.append(sentence)
        except:
            sentence = splits[n]
            sentence = re.sub("^\s*", "", sentence)
            if len(sentence) > 0:
                sentences.append(sentence)
            break
    return sentences
get_sentences(string)
```

```python
%%time
AGT["sentences"] = AGT["clean_string"].apply(get_sentences)
```

# Spacy NLP model

```python
nlp = spacy.load("../spacy_model")
ag_lemma_lookup = nlp.vocab.lookups.get_table("ag_lemma_lookup")
```

```python
# test whether our lookup table work
ag_lemma_lookup["VERB"]["μνήσομαι"]
```

```python

# !pip install greek-accentuation
from greek_accentuation.syllabify import rebreath
from greek_accentuation.accentuation import recessive, on_penult

def grave_to_acute(string):
    GRAVE = "\u0300"
    ACUTE = "\u0301"
    return unicodedata.normalize("NFC", "".join(unicodedata.normalize("NFD", string).replace(GRAVE, ACUTE)))

def play_with_morph(morph, tag):
    if isinstance(morph, str):
        try: 
            morph_lower = on_penult(rebreath(morph.lower()))
            lemma = ag_lemma_lookup[tag][morph_lower]
            return lemma
        except:
            try:
                morph_lower_cap = morph_lower.capitalize()
                lemma = ag_lemma_lookup[tag][morph_lower_cap]
                return lemma
            except: return morph
    else:
        return ""

def update_lemma(token_text, old_lemma, tag):
    if old_lemma == token_text:
        try:
            new_lemma = ag_lemma_lookup[tag][token_text] 
        except:
            try:
                new_lemma = ag_lemma_lookup[token.pos_][token.text.lower()] 
            except:
                new_lemma = play_with_morph(token_text, tag)
                if new_lemma == old_lemma:
                    new_lemma = play_with_morph(grave_to_acute(token_text), tag)
        return new_lemma
    else:
        return old_lemma

def update_lemmata(doc):
    for token in doc:
        token.lemma_ = update_lemma(token.text, token.lemma_, token.pos_)
    return doc
```

```python
def apply_nlp(sentences_list):
    spacy_docs = []
    for sentence in sentences_list:
        doc = update_lemmata(nlp(sentence))
        spacy_docs.append(doc)
    return spacy_docs
```

```python
from spacy.tokens import DocBin

def get_doc_bin_bytes(docs):
    """based on https://spacy.io/api/docbin"""
    doc_bin = DocBin()
    for doc in docs:
        doc_bin.add(doc)
    doc_bin_bytes = doc_bin.to_bytes()
    return doc_bin_bytes
```

```python
def get_lemmatized_sentences(docs):
    tags = ["NOUN", "VERB", "ADJ"]
    sents = []  
    for doc in docs:
        sent = []
        for token in doc:
            if token.pos_ in tags:
                sent.append(token.lemma_)
        sents.append(sent)
    return sents
```

```python
AGT["lemmatized_sentences"] = AGT["docs"].apply(get_lemmatized_sentences)
```

```python
def flat_lemmata(sentences):
    lemmata = [word for sent in sentences for word in sent]
    return lemmata
AGT["lemmata"] = AGT["lemmatized_sentences"].apply(flat_lemmata)
```
