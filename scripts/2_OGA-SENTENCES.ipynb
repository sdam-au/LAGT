```python
# Define file suffixes we're interested in
FILE_SUFFIXES = {
    "tokens": ".tok01.xml",
    "lemmas": ".tok01_lemma01.xml",
    "pos": ".tok01_postag01.xml",
    "cts": ".tok01_cts01.xml",
}


# Function to gather file paths from directories
def gather_file_paths(doc_dirs):
    file_structure = {}

    for subfolder in doc_dirs:
        subfolder_name = Path(subfolder).name  # Get the folder name
        file_structure[subfolder_name] = {}  # Create a key for the subfolder

        for file_type, suffix in FILE_SUFFIXES.items():
            # Search for files matching each suffix in the current subfolder
            matching_files = list(Path(subfolder).glob(f"*{suffix}"))

            if matching_files:
                # Use the first matching file (assuming one per type per subfolder)
                file_structure[subfolder_name][file_type] = str(matching_files[0])
            else:
                # No file found for this type; mark as None
                file_structure[subfolder_name][file_type] = None

    return file_structure
```

## Define functions for parsing the files data

```python
# keep only basic punctuation (no quotes/brackets)
_PUNCT = r".,;:!?"

# If you still keep a global class, do NOT include \s (we remove whitespace in the cleaner)
# Also include combining diacritics to be safe.
_CLASS = rf"[^\u0370-\u03FF\u1F00-\u1FFF\u0300-\u036F{re.escape(_PUNCT)}]"

# alternative
# _CLASS = rf"[^---\s{re.escape(_PUNCT)}]"

import re, unicodedata
from bs4 import BeautifulSoup, Comment
from pathlib import Path

def norm_clean_token(text: str) -> str:
    """NFC, drop controls+ALL whitespace, keep only Greek (incl. combining) + _PUNCT."""
    if not text:
        return ""
    s = unicodedata.normalize("NFC", text)
    # remove controls/format + ALL whitespace
    s = re.sub(r"[\u0000-\u001F\u007F-\u009F\u200B\u200C\u200D\u2060\uFEFF\s]+", "", s)
    # allow-list Greek + combining + chosen punctuation
    s = re.sub(_CLASS, "", s)
    return s

def paths_to_soups(doc_dir):
    try:
        path_tokens = str(next(Path(doc_dir).glob("*tok01.xml")))
        path_lemmas = str(next(Path(doc_dir).glob("*tok01_lemma01.xml")))
        path_pos    = str(next(Path(doc_dir).glob("*tok01_postag01.xml")))
        path_cts    = str(next(Path(doc_dir).glob("*tok01_cts01.xml")))
    except StopIteration:
        raise FileNotFoundError("One or more required files are missing in the directory.")

    with open(path_tokens, "r", encoding="utf-8") as f_tokens, \
         open(path_lemmas, "r", encoding="utf-8") as f_lemmas, \
         open(path_pos,    "r", encoding="utf-8") as f_pos, \
         open(path_cts,    "r", encoding="utf-8") as f_cts:
        soup_tokens = BeautifulSoup(f_tokens, "xml")
        soup_lemmas = BeautifulSoup(f_lemmas, "xml")
        soup_pos    = BeautifulSoup(f_pos,    "xml")
        soup_cts    = BeautifulSoup(f_cts,    "xml")

    return soup_tokens, soup_lemmas, soup_pos, soup_cts


POS_REVERSE = {
    "n": "NOUN",
    "v": "VERB",
    "a": "ADJ",
    "r": "ADP",     # but note: 'r' is often used for "adverbs" too in some corpora
    "p": "PRON",
    "l": "DET",     # articles
    "c": "CCONJ",   # or SCONJ; both mapped to 'c'
    "d": "ADV",
    "u": "PUNCT",
    "g": "PART",    # e.g. ,  etc.  particles
    "z": "X",       # uncategorized, foreign, artificial ellipsis etc.
}

def extract_token_data(soup_tokens, soup_lemmas, soup_pos, soup_cts):
    tokens_data = {}

    # tokens
    for mark in soup_tokens.find_all("mark"):
        token_id = mark.get("id")
        if not token_id:
            continue
        comment = mark.find_next(string=lambda t: isinstance(t, Comment))
        raw_tok = comment.strip() if comment else ""
        tokens_data[token_id] = {
            "token": raw_tok,
            "pos_index": extract_position(mark.get("xlink:href", "")),
        }

    # lemmas
    for feat in soup_lemmas.find_all("feat"):
        tid = feat.get("xlink:href", "").lstrip("#")
        if tid in tokens_data:
            tokens_data[tid]["lemma"] = feat.get("value", "")

    # POS
    for mf in soup_pos.find_all("multiFeat"):
        tid = mf.get("xlink:href", "").lstrip("#")
        if tid in tokens_data:
            pos_feature = mf.find("feat", {"name": "pos"})
            if pos_feature:
                tokens_data[tid]["pos"] = pos_feature.get("value", "")

    # CTS subunit
    for feat in soup_cts.find_all("feat"):
        tid = feat.get("xlink:href", "").lstrip("#")
        if tid in tokens_data:
            tokens_data[tid]["subunit"] = feat.get("value", "")

    # Combine & sort; drop items whose cleaned surface is empty
    token_tuples = sorted(
        [
            (
                (surf := norm_clean_token(data.get("token", ""))),
                norm_clean_token(data.get("lemma", "")),
                POS_REVERSE.get(data.get("pos", "")[:1], "X"),              # keep full; we'll slice to pos[0] in sentence stage
                data.get("subunit", ""),
                data.get("pos_index", (0, 0)),
            )
            for data in tokens_data.values()
            if (surf := norm_clean_token(data.get("token", "")))  # ensures surface not empty
        ],
        key=lambda x: x[4]
    )

    return token_tuples
```

```python
# --- robust xlink:href string-range parsing (kept as-is; used upstream) ---
_STRING_RANGE_RE = re.compile(
    r"""string-range\(
        [^,]*,                # xpath (ignore)
        \s*['"].*?['"]\s*,    # quoted string (ignore)
        \s*(\d+)\s*,          # start
        \s*(\d+)\s*           # length
    \)""",
    re.IGNORECASE | re.VERBOSE
)

def extract_position(href: str):
    """Return (start, length) from xlink:href with string-range(..., start, length); (0,0) if not parseable."""
    m = _STRING_RANGE_RE.search(href or "")
    if not m:
        return (0, 0)
    return (int(m.group(1)), int(m.group(2)))


# --- sentence segmentation + spacing rules (target shape) ---
# Sentence terminators to split on (include middle dot + ano teleia if they end sentences in OGA):
_SENT_END = {".", ";", "?", "!", "", ""}

# Punctuation that should NOT receive a leading space:
_NO_LEADING_SPACE = set(".,;:!?")

def extract_sentence_data(token_tuples):
    """
    Input token_tuples items: (surface, lemma, pos, subunit, (start, length))
    Returns: list of (sent_n, sentence_text, tokens_list), where tokens_list has:
             (surface, lemma, pos1, {'oga_cts': subunit}, start, end)
    NOTE: start/end are recomputed for the CLEANED surfaces with on-the-fly spacing.
    """
    sentences = []
    cur = []
    sent_n = 0

    for surface, lemma, pos, subunit, _ in token_tuples:
        cur.append((surface, lemma, pos, subunit))
        if surface in _SENT_END:
            sent_txt, toks = process_sentence(cur)
            sentences.append((sent_n, sent_txt, toks))
            sent_n += 1
            cur = []

    if cur:
        sent_txt, toks = process_sentence(cur)
        sentences.append((sent_n, sent_txt, toks))

    return sentences


def process_sentence(tokens):
    """
    tokens: list of (surface, lemma, pos, subunit)
    Build sentence text with one space between words, but NO space before punctuation in _NO_LEADING_SPACE.
    Produce tokens as (surface, lemma, pos1, {'oga_cts': subunit}, start, end),
    where start/end index into the constructed sentence_text.
    """
    pieces = []
    rec = []
    cursor = 0

    for i, (surface, lemma, pos, subunit) in enumerate(tokens):
        if not surface:
            continue

        # add a single space before non-punctuation tokens (not at sentence start)
        add_space = (i > 0) and (surface not in _NO_LEADING_SPACE)
        if add_space:
            pieces.append(" ")
            cursor += 1

        start = cursor
        pieces.append(surface)
        cursor += len(surface)
        end = cursor

        ref = {"oga_cts": subunit} if subunit else {}
        pos1 = (pos or "X")
        rec.append((surface, lemma or surface, pos1, ref, start, end))

    return "".join(pieces), rec
```

```python
from pathlib import Path
import re

_HAS_GREEK = re.compile(r"[---]")

doc_dir = "/srv/data/greek/OGA/oga/oga_v0.2.0_1/pta0041.pta001.pta-grc1"
doc_name = Path(doc_dir.rstrip("/")).name
doc_id = ".".join(doc_name.split(".")[:2])  # e.g., "pta0041.pta001"

soup_tokens, soup_lemmas, soup_pos, soup_cts = paths_to_soups(doc_dir)
token_tuples = extract_token_data(soup_tokens, soup_lemmas, soup_pos, soup_cts)

# triplets: (sent_n, sentence_text, tokens_list)
triplets = extract_sentence_data(token_tuples)

# Build, filter (must contain at least one Greek char), and reindex in one go
sentences_data = [
    (doc_id, i, sentence_text, tokens_list)
    for i, (orig_n, sentence_text, tokens_list) in enumerate(
        (t for t in triplets if any(_HAS_GREEK.search(tok[0]) for tok in t[2]))
    )
]
sentences_data[:3]
```

```python
%%time
for n, doc_dir in enumerate(doc_dirs):
    doc_id, sep, cts_source = doc_dir.rpartition("/")[-1].rpartition(".")
    out_fn = doc_id + ".pickle"

    if out_fn not in processed:
        soup_tokens, soup_lemmas, soup_pos, soup_cts = paths_to_soups(doc_dir)
        token_tuples = extract_token_data(soup_tokens, soup_lemmas, soup_pos, soup_cts)
        triplets = extract_sentence_data(token_tuples)  # (sent_n, text, tokens)

        # Filter out sentences with no Greek characters
        filtered = [
            (doc_id, i, text, toks, cts_source)
            for i, (orig_n, text, toks) in enumerate(
                t for t in triplets if any(_HAS_GREEK.search(tok[0]) for tok in t[2])
            )
        ]

        with open(os.path.join(target_path, out_fn), "wb") as f:
            pickle.dump(filtered, f)

        if n % 100 == 0:
            print(n)
```
