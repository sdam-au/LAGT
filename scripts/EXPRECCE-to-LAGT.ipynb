```python
import re
import os
import pandas as pd
```

```python
exprecce = pd.read_csv("../data/exprecce_v1.csv", sep=",")
exprecce["provencience"] = "christian"
exprecce["source"]  = "exprecce"
exprecce
```

```python
fn = "READY_JOANNES CHRYSOSTOMUS De sanctis martyribus {2062.050} (PG050.645-654) 20240513.txt"
rex = re.search("(\d{4})\.(\d{3})", fn).groups()
```

```python
exprecce_texts = []
source_dir = "../data/large_files/exprecce_data/"
for fn in os.listdir(source_dir):
    if ".txt" in fn:
        try:
            rex = re.search("(\d{4})\.(\d{3})", fn).groups()
            doc_id = "tlg{0}.tlg{1}".format(*rex)
            with open(source_dir + fn) as f:
                string = f.read()
            exprecce_texts.append({"doc_id": doc_id, "string": string})
        except:
            print(fn)    
```

```python
exprecce_texts_df = pd.DataFrame(exprecce_texts)
exprecce_texts_df
```

```python
exprecce.merge(exprecce_texts_df, on="doc_id")
```
